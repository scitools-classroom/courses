The following snippet of code priduces coordinate arrays and some data in a rotated pole coordinate system.
    
    import numpy as np
    import matplotlib.pyplot as plt
    import cartopy.crs as ccrs
    
    x = np.linspace(310, 390, 25)

    x2d, y2d = np.meshgrid(x, y)

1\. Define a cartopy coordinate reference system which represents a rotated pole with a pole longitude of 37.5 and a pole latitude of 177.5.

```python
rotated_pole = ccrs.RotatedPole(pole_longitude=177.5, pole_latitude=37.5)
```

2\. Produce a map, with coastlines, using the coordinate reference system created in #1.

```python
ax = plt.axes(projection=rotated_pole)
ax.coastlines()
plt.show()
```

3\. Produce a map, with coastlines, in a Plate Carree projection with a pcolormesh of the data generated by the code snippet provided at the beginning of the example - remember the data is in the rotated coordinate system defined in #1.

```python
x = np.linspace(310, 390, 25)
y = np.linspace(-24, 25, 35)
x2d, y2d = np.meshgrid(x, y)

data = np.cos(np.deg2rad(y2d) * 4) + np.sin(np.deg2rad(x2d) * 4)

ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()

plt.pcolormesh(x2d, y2d, data, transform=rotated_pole)

plt.show()
```
